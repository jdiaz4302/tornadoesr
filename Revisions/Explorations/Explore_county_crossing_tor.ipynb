```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

# 2016

For 2016, printing the narratives for all tornadoes when the narrative contains the word "border"

```python
tor_df = pd.read_csv('../Downloads/StormEvents_details-ftp_v1.0_d2016_c20180718.csv.gz')
tor_df = tor_df[tor_df['EVENT_TYPE'] == 'Tornado']
tor_df = tor_df.reset_index()
indicators = ['border' in nar for nar in tor_df['EVENT_NARRATIVE'].values]
for i in range(len(indicators)):
    if indicators[i] == True:
        print(i, '\n', tor_df['EVENT_NARRATIVE'][i], '\n')
```
